### 1.1 Drinking Fountains

```python
drinking_fountain_facility_columns = [
    'function', # 'Drinking Fountain'
    'latitude',
    'longitude',
    # optional
    'type',
    'info'
]
```

```python
dfs['drinking-fountains'].head()
```

```python
# read drinking-fountains dataframe
df = dfs['drinking-fountains']

# Helper functions:
# Split 'Co-ordinates' column into separate fields: 'latitude', 'longitude'
def split_coordinate(line):
    [lat, lon] = line.split(',')
    return lat, lon.strip()

# split 'Description' column into separate fields: type', 'info'
def split_description(line):
    # define indices for each field
    [i_type, i_subtype, i_info] = [1, 2, 3]
    
    # get fields
    parts = line.split(' - ')

    if len(parts) == 4:
        fountain_type = parts[i_type] + " / " + parts[i_subtype]
        return fountain_type, parts[i_info]
    elif len(parts) == 3:
        fountain_type = parts[i_type] + " / " + parts[i_subtype]
        return fountain_type, None
    elif len(parts) == 2:
        return parts[i_type], None
    else:
        return None, None

# data cleaning
df[['latitude', 'longitude']] = df['Co-ordinates'].apply(lambda x: pd.Series(split_coordinate(x)))
df[['type', 'info']] = df['Description'].apply(lambda x: pd.Series(split_description(x)))
df['category'] = 'drinking_fountains'

# Drop unwanted columns
clean_df = df.drop(columns=['Description', 'Co-ordinates', 'lat', 'lon'])
clean_df
```

```python
# Add clean data to faci_dfs
faci_dfs['drinking-fountains'] = clean_df
faci_dfs['drinking-fountains']
```

### 1.2 Toilet

```python
toilet_facility_columns = [
    'function', # 'Toilet'
    'latitude',
    'longitude',
    # optional
    'type',
    'address',
    'operator',
    'female',
    'male',
    'wheelchair',
    'baby_facil'
] 
```

```python
dfs['toilets'].head()
```

```python
# read drinking-fountains dataframe
df = dfs['toilets']
df
# Helper functions:
# split 'name' column into separate fields: 'type', 'address'
def split_name(line):
    # define indices for each field
    [i_type, i_address, i_detail_address] = [1, 2, 3]
    
    # get fields
    parts = line.split(' - ')
    
    if len(parts) == 4:
        address = parts[i_address] + parts[i_detail_address]
        return parts[i_type], address
    elif len(parts) == 3:
        return parts[i_type], parts[i_address]
    elif len(parts) == 2:
        return parts[i_type], None
    else:
        return None, None

# data cleaning
df[['type', 'address']] = df['name'].apply(lambda x: pd.Series(split_name(x)))
df['category'] = 'toilets'

# Drop unwanted columns
df.rename(columns={'lat': 'latitude'}, inplace=True)
df.rename(columns={'lon': 'longitude'}, inplace=True)
clean_df = df.drop(columns=['name'])
clean_df
```
